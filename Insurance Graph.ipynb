```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("insurance.csv")
df.head()
```

```python
plt.plot(df.age,df.bmi,color='g')
plt.xlabel("Age")
plt.ylabel("BMI")
plt.title("Insurance graph")
plt.show()
```

```python
plt.scatter(df.bmi,df.children,color='r')
plt.xlabel("BMI")
plt.ylabel("CHILDREN")
plt.title("Insurance graph")
plt.show()
```

```python
plt.bar(df.age,df.charges,color='b')
plt.xlabel("AGE")
plt.ylabel("CHARGES")
plt.title("Insurance graph")
plt.show()
```
